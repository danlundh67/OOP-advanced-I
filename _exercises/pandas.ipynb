Find out some statistical information about the salary, e.g. mean, median, min, max, 10 percentile and 90 percentile.

```python
print(f"genomsnittslön {df2['Salary_SEK'].mean()} median {df2['Salary_SEK'].median()}, min {df2['Salary_SEK'].min()}")

print(f"max {df2['Salary_SEK'].max()} 10% {df2['Salary_SEK'].quantile(0.10)}, 90% {df2['Salary_SEK'].quantile(0.9)}")
```

Group by different roles and take the median and average salaries for each group

```python
df2.info()
```

```python
tmp2=pd.DataFrame()
tmp2=df2.groupby(by='Position')['Salary_SEK'].median().reset_index()

tmp3=pd.DataFrame()
tmp3=df2.groupby(by='Position')['Salary_SEK'].mean().reset_index()

res1=pd.DataFrame()
res1=pd.merge(tmp2, tmp3, on='Position', how='inner')
res1.rename(columns={'Salary_SEK_x': 'median', 'Salary_SEK_y':'mean'})
```

Group by different departments and take the median and average salaries for each group

```python
tmp2=pd.DataFrame()
tmp2=df2.groupby(by='Department')['Salary_SEK'].median().reset_index()

tmp3=pd.DataFrame()
tmp3=df2.groupby(by='Department')['Salary_SEK'].mean().reset_index()

res2=pd.DataFrame()
res2=pd.merge(tmp2, tmp3, on='Department', how='inner')
res2.rename(columns={'Salary_SEK_x': 'median', 'Salary_SEK_y':'mean'})
```

b) Make a bar chart over the top five countries in medal count.


```python
df_gold=df_ol[12]

def plotbar(df, field, number):
    wdf=df.head(number).sort_values(by=field, ascending=False)
    wdf.plot.bar(x='NOC', y=field)


#df_gold.head(10).sort_values(by='Gold', ascending=False)

plotbar(df_gold, 'Total', 5)
```
